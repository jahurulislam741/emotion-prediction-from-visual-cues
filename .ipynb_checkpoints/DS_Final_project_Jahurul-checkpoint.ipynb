# Introduction to the problem
The problem at hand is to classify/predict speech emotions based on visual cues (WITHOUT any audio): movements of lips, jaw, eye-brows, etc.

## The dataset
A subset of videos from the [RAVDESS](https://zenodo.org/record/1188976) corpus, which contains video recordings representing different speech emotions (happy, angry, sad, etc.) performed by professional actors, was used to extract the Facial Movement Parameters or FAPs using an open-source toolkit named [OpenFace](https://github.com/TadasBaltrusaitis/OpenFace).  
To be more specific, when a video file is fed into OpenFace, it returns a number of "Action Units" (AUs) representing the magnitude of the movement of different parts of the face (e.g., lip corner pulling, lip raising/lowering, jaw-dropping, eye-brow raising, etc.) including eye gaze as a function of time. More information on the AUs can be found [HERE](https://github.com/TadasBaltrusaitis/OpenFace/wiki/Action-Units).

The raw output for each sound clip in the data for this project are prodided in the folder "1_speech_OpenFace_AUs"; the raw OpenFace output files contain a total of 714 columns. However, we are not going to use all the raw information OpenFace provides since only a few of them are relevant for our purpose; rather, we'll use only 17 AUs representing the lip and eye-brow actions. 

The final dataset to be used in this task contain 8 (eight) different emotion categories:
01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised. 

While the RAVDESS dataset also contains emotion in songs, our dataset come only from speech.

#### Extract Action Units (17 columns out of 714)

```python
# FUNCTION to extract AU columns and rename the columns
def extract_AU_cols_single_example(input_df):
    df = input_df.copy()
    df.columns = [c.strip() for c in df.columns] # remove white spaces from column names   

    # subset AU_r columns (representing the movements of lips, cheeks, jaw, and eye brows)
    cols = [c for c in df.columns if c.startswith('AU') and c.endswith('_r')] # keep only AU_r columns
    df = df[cols]
    colnames_full = ['inn_br_raise','out_br_raise','br_lower','up_lid_raise','cheek_raise',
                     'lid_tight','nose_wrinkle','up_lip_raise','lip_corner_pull','dimpler',
                     'lip_corner_dpress','chin_raise','lip_stretch','lip_tight','lips_part',
                     'jaw_drop','blink']

    df.columns = colnames_full
    df = df.T # transpose so that columns are timepoints

    return df
```

### Preprocessing of the raw data
First, we will preprocess the data. For each source data file, we'll:  
+ remove the columns except for the AUs relevent for us
+ change dimension of the matrix so that data from each video clip are matched for dimensions
    - we'll simply pad zeros equally on both sides of the matrix, where necessary
+ write processed matrices as back as CSV files in a separate folder

#### Match dimensions of each example by padding on both sides

```python
# function: pad 2D numpy array with zeros equally on left and right
def pad_zeros_on_left_and_right(np_array, pad_width):
    '''
    pad 2D numpy array with zeros equally on left and right
    '''
    padded_on_four_sides = np.pad(np_array, pad_width)
    padded_on_three_sides = padded_on_four_sides[pad_width:]
    padded_on_two_sides = padded_on_three_sides[:-pad_width]
    return padded_on_two_sides
```

```python
# get filenames of the data files
dir_path = "1_speech_OpenFace_AUs"
filenames = os.listdir(dir_path)
filenames = [f for f in filenames if ".csv" in f]

# get dimension of the longest data file
nrows_l = []
i = 0
for fname in filenames:
    clear_output(wait=True)
    print(f'Files processed: {i+1} of {len(filenames)}')
    i += 1
    fpath = os.path.join(dir_path, fname) # get full "relative" path of the file
    df = pd.read_csv(fpath) # read the file's data
    nrows_l.append(df.shape[0])
print(max(nrows_l))
```

```python
# read csv files and pad with zeros to match dimensions
# write each file back as CSV
desired_ncol = max(nrows_l)
# desired_ncol = 157

out_dir = '2_speech_AUs_dimension_matched'
i = 0
for fname in filenames[:]:
    clear_output(wait=True)
    i += 1
    print(f'Files processed: {i}')
    
    fpath = os.path.join(dir_path, fname) # get full "relative" path of the file
    df = pd.read_csv(fpath) # read the file's data
    df = extract_AU_cols_single_example(df)
    
    # convert to np array
    df = df.to_numpy() 

    # pad zeros on left and right side for df's having <157 timepoints
    pad_width = round((desired_ncol - df.shape[1])/2)
    df_padded = pad_zeros_on_left_and_right(df, pad_width)
    
    df_padded = df_padded[:, :(desired_ncol-3)] # drop rows from the end to match col numbers
    print(df_padded.shape)    

    df = pd.DataFrame(df_padded) # rows are action units; cols are timepoints
    
    # write to CSV
    out_fpath = os.path.join(out_dir, fname)
    df.to_csv(out_fpath, index = False) # write to CSV
```

### Build and run the model

```python
# define function to build CNN model
def build_model(input_shape):
    # build network topology
    model = keras.Sequential()

    # 1st conv layer
    model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(keras.layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same'))
    model.add(keras.layers.BatchNormalization())

    # 2nd conv layer
    model.add(keras.layers.Conv2D(32, (3, 3), activation='relu'))
    model.add(keras.layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same'))
    model.add(keras.layers.BatchNormalization())

    # flatten output and feed it into dense layer
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(64, activation='relu'))
    # model.add(keras.layers.Dropout(0.3))

    # output layer
    model.add(keras.layers.Dense(10, activation='softmax'))

    return model
```

```python
# function to plot training and testing accuracy across epochs
def plot_history(history):

    fig, axs = plt.subplots(2)

    # create accuracy sublpot
    axs[0].plot(history.history["accuracy"], label="train accuracy")
    axs[0].plot(history.history["val_accuracy"], label="val accuracy")
    axs[0].set_ylabel("Accuracy")
    axs[0].legend(loc="lower right")
    axs[0].set_title("Accuracy eval")

    # create error sublpot
    axs[1].plot(history.history["loss"], label="train error")
    axs[1].plot(history.history["val_loss"], label="val error")
    axs[1].set_ylabel("Error")
    axs[1].set_xlabel("Epoch")
    axs[1].legend(loc="upper right")
    axs[1].set_title("Error eval")
    
    plt.tight_layout()
    plt.show()
```

```python
from tensorflow import keras

# create network
input_shape = (X_train.shape[1], X_train.shape[2], 1)
model = build_model(input_shape)

# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=optimiser,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# train model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs= 30)
```

```python
# plot accuracy/error for training and validation
plot_history(history)
```
